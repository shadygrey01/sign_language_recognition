```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from sklearn.metrics import roc_auc_score
import numpy as np

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
split_dir = "split_dataset/"

# Define data transforms
data_transforms = {
    "train": transforms.Compose([
        transforms.Resize((224, 224)),  # ConvNeXt expects 224x224
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # ImageNet norms
    ]),
    "val": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    "test": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
}

# Load datasets
image_datasets = {
    split: datasets.ImageFolder(
        os.path.join(split_dir, split),
        transform=data_transforms[split]
    ) for split in ["train", "val", "test"]
}

# Create data loaders
dataloaders = {
    split: DataLoader(
        image_datasets[split],
        batch_size=32,
        shuffle=(split == "train"),
        num_workers=4
    ) for split in ["train", "val", "test"]
}

# Dataset sizes and class names
dataset_sizes = {split: len(image_datasets[split]) for split in ["train", "val", "test"]}
class_names = image_datasets["train"].classes  # ['mpox', 'cpox', 'measles', 'normal']
num_classes = len(class_names)
print(f"Dataset sizes: {dataset_sizes}")
print(f"Classes: {class_names}")

# Load pre-trained ConvNeXtXLarge
model = models.convnext_xlarge(weights=models.ConvNeXt_XLarge_Weights.IMAGENET1K_V1)
model = model.to(device)

# Freeze lower layers (feature extractor)
for name, param in model.named_parameters():
    if "features" in name and int(name.split(".")[1]) < 6:  # Freeze first 6 stages (out of 8)
        param.requires_grad = False

# Modify the classifier for 4 classes
in_features = model.classifier[2].in_features  # Get input features of the final layer
model.classifier[2] = nn.Linear(in_features, num_classes)  # Replace with 4-class output
model.classifier = model.classifier.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)

# Training function
def train_model(model, criterion, optimizer, num_epochs=10):
    best_acc = 0.0
    best_model_wts = model.state_dict()

    for epoch in range(num_epochs):
        print(f"Epoch {epoch+1}/{num_epochs}")
        print("-" * 10)

        # Each epoch has a training and validation phase
        for phase in ["train", "val"]:
            if phase == "train":
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0
            all_labels = []
            all_probs = []

            # Iterate over data
            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                # Forward pass
                with torch.set_grad_enabled(phase == "train"):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == "train":
                        loss.backward()
                        optimizer.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

                # For AUC
                probs = torch.softmax(outputs, dim=1).detach().cpu().numpy()
                all_labels.extend(labels.cpu().numpy())
                all_probs.extend(probs)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]
            
            # Compute AUC (multi-class: one-vs-rest)
            all_labels = np.array(all_labels)
            all_probs = np.array(all_probs)
            auc_scores = [roc_auc_score(all_labels == i, all_probs[:, i]) for i in range(num_classes)]
            avg_auc = np.mean(auc_scores)

            print(f"{phase.capitalize()} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} AUC: {avg_auc:.4f}")

            # Save best model
            if phase == "val" and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = model.state_dict()

        print()

    print(f"Best val Acc: {best_acc:.4f}")
    model.load_state_dict(best_model_wts)
    return model

# Test function
def evaluate_model(model):
    model.eval()
    running_corrects = 0
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for inputs, labels in dataloaders["test"]:
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            running_corrects += torch.sum(preds == labels.data)
            probs = torch.softmax(outputs, dim=1).cpu().numpy()
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs)

    test_acc = running_corrects.double() / dataset_sizes["test"]
    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)
    auc_scores = [roc_auc_score(all_labels == i, all_probs[:, i]) for i in range(num_classes)]
    avg_auc = np.mean(auc_scores)

    print(f"Test Acc: {test_acc:.4f} Test AUC: {avg_auc:.4f}")

# Train the model
model = train_model(model, criterion, optimizer, num_epochs=10)

# Save the model
torch.save(model.state_dict(), "convnext_xlarge_finetuned.pth")

# Evaluate on test set
evaluate_model(model)
```

```python
import os
import torch
import torchvision.models as models
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from sklearn.metrics import roc_auc_score
import numpy as np

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
split_dir = "split_dataset/"
model_path = "convnext_large_finetuned.pth"

# Define data transforms (same as training)
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # ConvNeXt expects 224x224
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # ImageNet norms
])

# Load test dataset
test_dataset = datasets.ImageFolder(os.path.join(split_dir, "test"), transform=data_transforms)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

# Dataset info
dataset_size = len(test_dataset)
class_names = test_dataset.classes  # ['cpox', 'measles', 'mpox', 'normal']
num_classes = len(class_names)
print(f"Test dataset size: {dataset_size}")
print(f"Classes: {class_names}")

# Load fine-tuned ConvNeXtLarge
model = models.convnext_large(weights=None)  # Initialize without pre-trained weights

# Modify the classifier to match the fine-tuned model (4 classes)
in_features = model.classifier[2].in_features  # Should be 1536 for ConvNeXtLarge
model.classifier[2] = torch.nn.Linear(in_features, num_classes)  # Replace with 4-class output
model = model.to(device)

# Load the fine-tuned weights
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()  # Set to evaluation mode

# Prediction function
def get_predictions(model, dataloader):
    all_labels = []
    all_probs = []
    all_preds = []
    all_paths = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)  # Convert logits to probabilities
            _, preds = torch.max(probs, 1)  # Get predicted class indices

            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            # Store image paths
            batch_paths = [dataloader.dataset.samples[i][0] for i in range(len(labels))]
            all_paths.extend(batch_paths)

    return np.array(all_labels), np.array(all_probs), np.array(all_preds), all_paths

# Get predictions
true_labels, probs, preds, image_paths = get_predictions(model, test_loader)

# Print predictions for each sample
print("\nPredictions on test set:")
print(f"{'Image Path':<50} {'True Label':<10} {'Predicted Label':<15} {'Probability':<15}")
print("-" * 90)
for i in range(len(true_labels)):
    true_class = class_names[true_labels[i]]
    pred_class = class_names[preds[i]]
    max_prob = probs[i][preds[i]]
    path = image_paths[i]
    print(f"{path:<50} {true_class:<10} {pred_class:<15} {max_prob:.4f}")

# Calculate metrics
accuracy = np.mean(preds == true_labels)
print(f"\nTest Accuracy: {accuracy:.4f}")

# Multi-class AUC (one-vs-rest)
auc_scores = [roc_auc_score(true_labels == i, probs[:, i]) for i in range(num_classes)]
avg_auc = np.mean(auc_scores)
print(f"Test AUC (average): {avg_auc:.4f}")
for i, class_name in enumerate(class_names):
    print(f"AUC for {class_name}: {auc_scores[i]:.4f}")
```

```python
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import numpy as np

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
model_path = "convnext_large_finetuned.pth"
image_path = "mes.jpeg"  # Replace with your image path

# Define data transforms (same as training)
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # ConvNeXt expects 224x224
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # ImageNet norms
])

# Load and preprocess the image
image = Image.open(image_path).convert("RGB")  # Ensure image is in RGB format
input_tensor = data_transforms(image)  # Apply transforms
input_batch = input_tensor.unsqueeze(0)  # Add batch dimension: [1, 3, 224, 224]

# Define class names (from your dataset)
class_names = ['cpox', 'measles', 'mpox', 'normal']
num_classes = len(class_names)

# Load fine-tuned ConvNeXtLarge
model = models.convnext_large(weights=None)  # Initialize without pre-trained weights

# Modify the classifier to match the fine-tuned model (4 classes)
in_features = model.classifier[2].in_features  # Should be 1536 for ConvNeXtLarge
model.classifier[2] = torch.nn.Linear(in_features, num_classes)  # Replace with 4-class output
model = model.to(device)

# Load the fine-tuned weights
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()  # Set to evaluation mode

# Move input to device
input_batch = input_batch.to(device)

# Make prediction
with torch.no_grad():
    output = model(input_batch)
    probs = torch.softmax(output, dim=1)  # Convert logits to probabilities
    predicted_prob, predicted_idx = torch.max(probs, 1)  # Get max probability and class index

# Convert to CPU and numpy for display
probs = probs.cpu().numpy()[0]  # [4]
predicted_prob = predicted_prob.cpu().numpy()[0]  # Scalar
predicted_idx = predicted_idx.cpu().numpy()[0]  # Scalar
predicted_class = class_names[predicted_idx]

# Display results
print(f"\nPrediction for image: {image_path}")
print(f"Predicted Class: {predicted_class}")
print(f"Predicted Probability: {predicted_prob:.4f}")
print("\nProbability Distribution:")
for i, class_name in enumerate(class_names):
    print(f"{class_name}: {probs[i]:.4f}")
```
